```python
import matplotlib.pyplot as plt
from dna_features_viewer import GraphicFeature, GraphicRecord
```

```python
fig, ax = plt.subplots(figsize=(10,5))
features=[
    GraphicFeature(start=0, end=20, strand=+1, color="#ffd700",
                   label="Small feature"),
    GraphicFeature(start=20, end=500, strand=+1, color="#ffcccc",
                   label="Gene 1 with a very long name"),
    GraphicFeature(start=20, end=500, strand=+1, color="#ffcccc",
                   label="Gene 1 with a very long name"),
    GraphicFeature(start=20, end=500, strand=+1, color="#ffcccc",
                   label="Gene 1 with a very long name"),
    GraphicFeature(start=400, end=700, strand=-1, color="#cffccc",
                   label="Gene 2.1"),
    GraphicFeature(start=400, end=700, strand=-1, color="#cffccc",
                   label="Gene 2.2"),
    GraphicFeature(start=600, end=900, strand=+1,
                   label="Gene 3.1"),
    GraphicFeature(start=600, end=900, strand=+1,
                   label="Gene 3.2")
]
for i in range(10):
    features.append(
        GraphicFeature(start=600, end=900, strand=+1, color="#ccccff",
                   label="Gene 3.2")
    )
record = GraphicRecord(sequence_length=1000, features=features*10)
record.plot(ax = ax, with_ruler=False, draw_line=False)
```

```python
DATA_DIR = f"../tests/test_data"
test_interval = "chr9:4000000-6000000"
test_itv = test_interval.replace(':', '_').replace('-', '_')

frame = XAxis() + \
    Cool(f"{DATA_DIR}/cool_{test_itv}.mcool") + \
    Spacer(1) + \
    Arcs(f"{DATA_DIR}/arcs_{test_itv}.arcs") + Inverted() + \
    GTF(f"{DATA_DIR}/gtf_{test_itv}.gtf", length_ratio_thresh=0.005) + TrackHeight(7) + \
    Spacer(1) + \
    BigWig(f"{DATA_DIR}/bigwig_{test_itv}.bw") + \
    Spacer(1) + \
    Bed(f"{DATA_DIR}/bed_{test_itv}.bed") + TrackHeight(10)

bsr = Browser(frame)
bsr.goto(test_interval)
bsr.show()
```
