```python
data_dir = "../../../tests/test_data/"
bw1 = f"{data_dir}/bigwig_chr9_4000000_6000000_K562_H3K27ac.bigwig"
bw2 = f"{data_dir}/bigwig_chr9_4000000_6000000_K562_H3K27me3.bigwig"
bw3 = f"{data_dir}/bigwig_chr9_4000000_6000000_K562_H3K4me3.bigwig"
bw4 = f"{data_dir}/bigwig_chr9_4000000_6000000_K562_RNA.bigwig"
```

```python
frame = \
    BigWig(bw4, alpha=0.6) + Color("#ff9c9c") + \
    BigWigCoverage(bw1, alpha=0.6, color="#fcbe07") + \
    Title("RNA / H3K27ac") + \
    BigWig(bw4, alpha=0.6) + Color("#ff9c9c") + \
    BigWigCoverage(bw3, alpha=0.6, color="#07fcbe") + \
    Title("RNA / H3K4me3") + \
    BigWig(bw4, alpha=0.6) + Color("#ff9c9c") + \
    BigWigCoverage(bw2, alpha=0.6, color="#66ccff") + \
    Title("RNA / H3K27me3")

frame *= MaxValue(20)
frame *= MinValue(0)
frame * TrackHeight(4)

frame = XAxis() + frame
# frame.plot("chr1:1-5000000")
frame.plot("chr9:4000000-6000000")
```
