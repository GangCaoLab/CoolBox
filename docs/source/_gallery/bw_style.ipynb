BigWig hist styles: 
+ fill
+ line
+ points

```python
bw = "../../../tests/test_data/bigwig_chr9_4000000_6000000.bw"

frame = XAxis() + \
    BigWig(bw) + \
    BigWig(bw) + Color("#fcbe07") + HistStyle("line", 1) + \
    BigWig(bw) + Color("#000000") + HistStyle("points", 1)

frame.plot("chr9:4700000-4950000")
```
